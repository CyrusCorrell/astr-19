```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n=100
x = np.zeros(n,dtype='float')
for i in range(n):
    x[i] = np.random.random() 
width = 0.5
hmin = np.floor(np.min(x))
hmax = np.ceil(np.max(x))
bins = np.arange(hmin,hmax,width)

plt.hist(x, bins=100, edgecolor='green')

```
